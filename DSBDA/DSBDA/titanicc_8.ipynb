```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
# Display first few rows of the dataset
print(titanic.head())
```

```python
# Set the style for the plots
sns.set_style("whitegrid")
```

```python
# 1. Survival countplot
plt.figure(figsize=(6,4))
sns.countplot(x="survived", hue="survived", data=titanic, palette="coolwarm", legend=False)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="survived", data=titanic, palette="coolwarm")
plt.title("Survival by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# 3. Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.countplot(x="pclass", hue="survived", data=titanic, palette="coolwarm")
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# 4. Age Distribution of Passengers
plt.figure(figsize=(8,5))
sns.histplot(titanic["age"].dropna(), kde=True, bins=30, color="purple")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
# 5. Survival based on Fare Paid
plt.figure(figsize=(8,5))
sns.boxplot(x="survived", y="fare", hue="survived", data=titanic, palette="coolwarm", legend=False)
plt.title("Survival based on Fare Paid")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
```

```python
#create boxplot 
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, palette='Set2')
```

```python
# Labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender and Survival Status')
plt.legend(title="Survived", labels=["No (0)", "Yes (1)"])

# Show the plot
plt.show()
```
